```python
file = open('todas_empresas.csv', 'r', encoding="utf8")
```

```python
lines = file.readlines()
```

```python
file.close()
```

```python
lines[1].replace('\n', '').split(';')
```

```python
empresas = []
for line in lines:
    splited = line.replace('\n', '').split(';')
    if len(splited) == 1:
        categoria_atual = splited[0]
    else:
        empresas.append({'empresa': splited[0], 'codigo': splited[1], 'setor': categoria_atual})
```

```python
empresas
```

```python
import pandas as pd
import yfinance as yf
from pandas_datareader import data as pdr
from datetime import datetime, date, timedelta
yf.pdr_override()
```

```python
start_date = (datetime.now() - timedelta(1)).strftime("%Y-%m-%d")
end_date = date.today().strftime("%Y-%m-%d")
```

```python
for empresa in empresas:
    opcoes = ['{}3.SA'.format(empresa['codigo']),
             '{}4.SA'.format(empresa['codigo']),
             '{}11.SA'.format(empresa['codigo'])]
    for opcao in opcoes:
        a = pdr.get_data_yahoo(opcao, start=start_date, end=end_date)
        if len(a) > 0 and a.iloc[0]['Volume'] > 0:
            empresa['codigo_yahoo'] = opcao
            empresa['codigo_b3'] = opcao.replace('.SA', '')
            break
```

```python
empresas
```

```python
f = open("empresas_b3.csv", "a")
f.write('Empresa;Código;Código Yahoo;Setor\n')
for empresa in empresas:
    if 'codigo_b3' in empresa:
        f.write('{};{};{};{}\n'.format(empresa['empresa'], empresa['codigo_b3'], empresa['codigo_yahoo'], empresa['setor']))
    else:
        f.write('{};{};{};{}\n'.format(empresa['empresa'], empresa['codigo'], empresa['codigo'], empresa['setor']))
f.close()
```
